## Clean data
You can review the data to find errors or problems. You can use the instruction 'df.head()' to get a first view of the data. 
In this example, we find the errors:
- Cells without any data
- Asterisks in the name column
- MonthYear column has time which is not required

```python
#Lets replace the cells without any data in the Name column with the value â€˜Noneâ€™ using this line
df = df.astype(object).where(pd.notnull(df),None)
df.head()
```

```python
#To remove the asterisk in the Name column.
df['Name']=df['Name'].apply(lambda x:str(x).replace('*',''))
df.head()
```

```python
#Firs step is to convert the data frame object to date.time format and then replace the date time with just date.
df['MonthYear'] = pd.to_datetime(df['MonthYear'])
df['MonthYear'] = df['MonthYear'].apply(lambda x: x.date())
df.head()
```

## Save the clean data

```python
df.to_csv(cleanexamplefile_path)
```
